# Question 3.2
### Importing data

```python
train = pd.read_csv('MIT801 Table 4.csv')
test = pd.read_csv('MIT801 Table 5.csv')
```

### Seperating target variable from data set

```python
X_train = train.iloc[:, 1:-1].values
y_train = train.iloc[:, -1].values
X_test = test.iloc[:, 1:-1].values
y_test = test.iloc[:, -1].values
```

### Predict label and return k-nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
x1 = [5.03, -4.97, 3.50]
x2 = [-3.11, -6.84, 10.71]
x3 = [-1.62, -6.34, 4.66]

classifier = KNeighborsClassifier(n_neighbors =1)
classifier.fit(X_train, y_train)
q1a = classifier.predict([x1])
q1b =  classifier.kneighbors([x1])
      
      
classifier = KNeighborsClassifier(n_neighbors =3)
classifier.fit(X_train, y_train)
q2a = classifier.predict([x2])
q2b =  classifier.kneighbors([x2])


classifier = KNeighborsClassifier(n_neighbors =5)
classifier.fit(X_train, y_train)
q3a = classifier.predict([x3])
q3b =  classifier.kneighbors([x3])
```
