# Question 2.1
### Importing the data

```python

data = {'x1':  [2.20, 4.23, 1.91, 2.19, 1.92, 1.67, 3.84, 2.34, 2.34, 1.63],
        'x2': [2.20, 3.64, 2.07, 2.12, 1.82, 1.49, 3.89, 1.66, 1.78, 1.86]
        }
X = pd.DataFrame (data, columns = ['x1','x2'])
X= X.to_numpy()
```

### Calculating the mean vector

```python
A = [2.20, 4.23, 1.91, 2.19, 1.92, 1.67, 3.84, 2.34, 2.34, 1.63]
B = [2.20, 3.64, 2.07, 2.12, 1.82, 1.49, 3.89, 1.66, 1.78, 1.86]

mv = [np.mean(A), np.mean(B)]

Y = X - mv
```

### Calculating the covariance matrix

```python
covMatrix = np.cov(np.transpose(X))
```

### Calculating the eigenvalues and eigenvectors

```python
eigval,eigvect= np.linalg.eig(covMatrix)
```

### Seperating eigenvectors into p1 and p2 vectors

```python
p1 = np.array([0.733691, 0.67948328])
p2 = np.array([-0.67948328, 0.733691])
```

### Calculating projection coefficients

```python
w1 = np.matmul(Y, -p1)
w2 = np.matmul(Y, p2)
```

### Manually classifying anomalies 

```python
c = []
for i in w1:
    if i<0:
        c.append('Anomaly')
    else:
        c.append('Normal')


data = {'w1': w1,
        'w2': w2,
        'Class': c
       }
principalDf1 = pd.DataFrame(data
             , columns = ['w1', 'w2', "Class"])
```

```python
print('')
print('')
print("Mean Vector: ", mv)
print('')
print("Covariance Matrix: ")
print(covMatrix)
print('')
print("Eigenvalues: ",eigval)
print('')
print('Eigenvectors for eigenvalue = 1.42: ', p1)
print('')
print('Eigenvectors for eigenvalue = 0.05: ', p2)
print('')
print("Principal Component Matrix(P): \n", eigvect)
print('')
print("Projection Coefficients (using negative sign of p1): \n", principalDf1)
print('')
print('')
```

### Functions to return the variable terms

```python
def gaussianv(mu, sig, row):
    detSig = np.linalg.det(sig)
    invSig = np.linalg.inv(sig)
    col = np.transpose(row)
    
    constant = 1/((2*np.pi)*(detSig**0.5))
    
    t1 = -0.5*np.matmul(row, invSig)
    exponent = np.matmul(t1, col)
    
    ans = constant*np.exp(exponent)
    
    return exponent

def gaussian(mu, sig, row):
    detSig = np.linalg.det(sig)
    invSig = np.linalg.inv(sig)
    col = np.transpose(row)
    
    constant = 1/((2*np.pi)*(detSig**0.5))
    
    t1 = -0.5*np.matmul(row, invSig)
    exponent = np.matmul(t1, col)
    
    ans = constant*np.exp(exponent)
    
    return ans
```
